```python
#Task 2:
import pandas as pd
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to resolve LookupError for 'punkt_tab'
print(df['customer_message'].head())

stop_words = set(stopwords.words('english'))
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove special characters, numbers, punctuation
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df['clean_text'] = df['customer_message'].apply(clean_text)
print(df[['customer_message', 'clean_text']].head())

df['tokens'] = df['clean_text'].apply(word_tokenize)
print(df['tokens'].head())

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]
df['filtered_tokens'] = df['tokens'].apply(remove_stopwords)
print(df['filtered_tokens'].head())

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Join tokens back into text
processed_text = df['filtered_tokens'].apply(lambda x: " ".join(x))

tokenizer = Tokenizer(num_words=5000)
tokenizer.fit_on_texts(processed_text)
sequences = tokenizer.texts_to_sequences(processed_text)
padded_sequences = pad_sequences(
    sequences,
    maxlen=20,
    padding='post',
    truncating='post'
)
print("Shape of padded sequences:", padded_sequences.shape)
print(padded_sequences[:2])
```

#### Q1.  
Recurrent Neural Network processes words one at a time by passing information from previous time steps. However, when sequences become long, early information gradually becomes weak due to the vanishing gradient problem. Because of this, RNNs often struggle to remember important words or context from earlier parts of a sentence.

#### Q2.
Long Short-Term Memory was designed to solve this limitation of RNNs.
LSTMs use special memory cells and gates:  
  * Forget gate
  * Input gate
  * Output gate

These gates help the model decide:  
  * What information to keep
  * What information to remove
  * What information to use later

Because of this, LSTMs can remember important information for longer sequences and perform better in NLP tasks.

#### Q3.
The Attention Mechanism allows the model to focus on the most relevant words in the input sequence while generating output.
Instead of depending only on the final hidden state, attention gives access to all important words.

Example:
In language translation:

Input:
“The payment issue was resolved quickly.”

When generating the translated word for payment, attention focuses directly on the word payment rather than trying to remember it indirectly.

This improves performance in tasks like:
* Machine translation
* Text summarization
* Question answering

#### Q4.
Transformer models are important because they replace recurrence with self-attention.

Advantages of transformers:
* Process all words in parallel
* Capture long-range dependencies efficiently
* Train faster on large datasets
* Understand contextual relationships better

Transformers are the foundation of modern Generative AI systems such as:
* GPT
* BERT

They are widely used for:
* Chatbots
* Text generation
* Translation
* Search
* Summarization
* Code generation
